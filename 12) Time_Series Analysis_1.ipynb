# Retrieve monthly data

```python
#df.Close  means select specific column(close) of the data frame

#resample('M') means resampling on a monthly frequency where 'M' means end date of the month 
            # for example July is for 31 and february is for 28
            #M for month end frequency
    
#mean() then to take average price

df.Close.resample('M').mean()
```

# To plot the chart of monthly data using matplotlib

```python
%matplotlib inline
df.Close.resample('M').mean().plot()
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot(kind="pie")
```

```python
%matplotlib inline
df.Close.resample('M').mean().plot(kind="bar")
```
