```python
import pandas as pd
```

```python
df=pd.read_csv("aapl_no_dates.csv")
df.head()
```

# to enter date using date_range()


# first create date range

```python
rng=pd.date_range(start="6/1/2017", end="6/30/2017", freq="B")
rng      #freq=B means business days frequency
```

# set it in index

```python
df.set_index(rng, inplace=True)
df
```

```python
%matplotlib inline
df.Close.plot()
```

```python
df["2017-06-01":"2017-06-12"]
```

```python
df["2017-06-01":"2017-06-12"].Close.mean()
```

```python
df["2017-06-01":"2017-06-12"].Volume.mean()
```

# asfreq('D') will allow you to regenerate your dataframe according to that new frequency

```python
 df.asfreq("D", method="pad")  #D means daily frequency
    
#latest weekday's data will be carry forward to the next weekends date
```

```python
df.asfreq("W", method="pad")
```

```python
df1=df.asfreq("W", method="pad")
df1
```

```python
%matplotlib inline
df1.Close.plot()
```

```python
%matplotlib inline
df1.Close.plot(kind="pie")
```

```python
df.asfreq("H", method="pad")
```
