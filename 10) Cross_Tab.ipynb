```python
import pandas as pd
```

```python
df=pd.read_excel("survey.xlsx")
df
```

 # crosstab is used to express frequency distribution
 
 # it will deal with only those elements that are mentioned in the function
 
 # first name will be in the x-axis and 2nd name will be in the y-axis

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Nationality,df.Sex)
```

```python
pd.crosstab(df.Nationality,df.Age)
```

```python
pd.crosstab(df.Sex,df.Age)
```

```python
pd.crosstab(df.Age,df.Name)
```

```python
pd.crosstab(df.Nationality,df.Sex)
```

```python
pd.crosstab(df.Nationality,df.Sex,margins=True) #margins will calculate the total numbers in x-axis and y-axis.
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Handedness],margins=True)
```

```python
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

# to print percentage nornamlize="index" function is used

```python
pd.crosstab([df.Sex],[df.Handedness], normalize='index') #it will print the percentage
```

```python
import numpy as np
pd.crosstab([df.Sex],[df.Handedness], values=df.Age, aggfunc=np.average)
```

```python
pd.crosstab([df.Sex],[df.Nationality], values=df.Age, aggfunc=np.average)
```
