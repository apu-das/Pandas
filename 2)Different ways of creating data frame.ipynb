```python
##if no people name then to convert NaN to any name s.t. sam walter 

import pandas as pd

def convert_people_cells(cell):
    if cell=="n.a.":
        return "sam walton"
    return cell
df=pd.read_csv('stock_data.csv', converters={
    'people': convert_people_cells})
df
```

```python
df
```

```python
##covert 'not available' to 'none' in eps cells.
def convert_eps_cells(cell):
    if cell=="not available":
        return "none"
    return cell
df=pd.read_csv('stock_data.csv', converters={
    'eps': convert_eps_cells})
df
```

```python
#create a new excel file

df.to_excel("new.xlsx",sheet_name="stocks")
```

```python
#Create two different data frame in a single file that contains two sheets
#one sheet named stocks another is weather

df_stocks=pd.DataFrame({
    'tickers': ['google','WMT','MSFT'],
    'price': [373,54,33],
    'pe': [40,43,55],
    'eps': [34,4,5]
    
})
df_weather=pd.DataFrame({'day':['16/09/2023','17/09/2023','18/09/2023'], 
             'temperature':[32,35,19],
             'windspeed':[6,7,2],
             'event':['Rain','sunny','snow']})


with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer,sheet_name='stocks',index=False) #Index number won't be printed
    df_weather.to_excel(writer,sheet_name='weather')  #Index number will be printed
    
```
