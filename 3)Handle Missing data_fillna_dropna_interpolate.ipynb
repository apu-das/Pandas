
# to set day as index


```python
import pandas as pd
df=df=pd.read_csv("weather_data1.csv", parse_dates=['day'])
df.set_index('day',inplace=True)
df
```


# fill na with zero(0)~ ~All NaN will be zero


```python
new_df= df.fillna(0)
new_df
```

```python
 #Not filling event by zero(0). "event" will be filled by "no event"
    
new_df=df.fillna({
    'temperature': 0,
    'windspeed':0,
    'event': 'no event'
})
new_df
```


# Forward and backward data carry forward
# Horizontally


```python
#temperature can't be zero within a day from 32.0. So if we carry previous temperature(32.0) to the next day then it'd look better. To do this we will use carry forward 

#carry forward= df.fillna(method="ffill").

new_df= df.fillna(method="ffill")
new_df
```

```python
#backward forward will print next day data to the previous day.
#carry backrward= df.fillna(method="bfill").

new_df= df.fillna(method="bfill")
new_df
```


# fill data vertically(column to column)


```python
new_df= df.fillna(method="bfill",axis="columns")
new_df
```


# using limit (one data will be print just one time and the remaining "NaN" will be as same as before. 


```python
new_df= df.fillna(method="ffill",limit=1)
new_df
```

```python
new_df= df.fillna(method="ffill",limit=2)
new_df
```

```python
new_df= df.fillna(method="ffill",limit=3)
new_df
```

# Interpolation (in between previous and next value)

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method="nearest")
new_df
```

```python
new_df=df.interpolate(method="cubic")
new_df
```

```python
new_df=df.interpolate(method="zero")
new_df
```

```python
new_df=df.interpolate(method="quadratic")
new_df
```

```python
new_df=df.interpolate(method="index")
new_df
```

```python
new_df=df.interpolate(method="time")  
new_df


#Time method is very powerfull. It looks about the time when making an interpolation
#The value is printed relating to the nearest time.
```


# Drop the rows which has NaN

```python
#drop the rows which has two NaN

new_df=df.dropna(thresh=2)
new_df
```
