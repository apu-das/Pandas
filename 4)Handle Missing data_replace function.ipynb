```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('weather_data_1.csv')
df
```

# Convert "0" to "no event" in event column

```python
def event_fill(cell):
    if cell=='0':
        return "no event"
    return cell
df=pd.read_csv('weather_data_1.csv',converters={
    'event': event_fill
})
df
```

# replace any value 

```python
#replace -99999 by NaN

new_df=df.replace(-99999,np.NaN)
new_df
```

```python
#replace to or more value using list

new_df=df.replace([-99999,-88888],np.NaN)
new_df
```

```python
#Using Dictionary

new_df=df.replace({
    'temperature': [-99999,-88888],
    'windspeed': -99999,
    'event': 'no event'
},np.NaN)
new_df
```

```python
new_df=df.replace({
    -99999: np.NaN,
    -88888: np.NaN,
    'no event': 'sunny'
})
new_df
```

# Regex() 
# removing any alphabet from any column

```python
new_df=df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},' ' ,regex=True)
new_df
```
