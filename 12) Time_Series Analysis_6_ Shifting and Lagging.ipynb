```python
import pandas as pd
df=pd.read_csv('fb.csv',parse_dates=['Date'], index_col='Date')
df
```

# shift() ~ it will just shift the data point to right hand side or forward direction

```python
# To shift price 171.00(first one) to one day later
#and the first one becomes NaN

df.shift(1)
```

```python
# To shift price 171.00(first one) to two day later
#and the first two become NaN
df.shift(2)
```

# shift(-ve) ~ it will just shift the data point to left hand side or reverse

```python
# To shift price 170.00(first one) to one day before and the last one becomes NaN

df.shift(-1)
```

```python
df.shift(-5)

#last five become NaN
```

# uses of shift() is to see the price change in one day or else

```python
df['Previous_day_price']= df['Price'].shift(1)
df

#creating new price column named "Previous_day_price" which takes the value of shift(1) function 
```

```python
# to calculate the change in percentage

df["1_day_change"]= df['Price']-df['Previous_day_price']
df
```

# getting 5 days percentage return

```python
df['5_day % _ return'] = ( df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```
